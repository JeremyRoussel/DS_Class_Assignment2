```python
import math
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import numpy as np
import pandas as pd
from sklearn.datasets import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Creates 3 joint probability centers with Gaussian distribution, with different variance for each peak.

mu1 = .5
mu2 = 1
mu3 = .75

center1x = np.random.normal(2 , mu1 , 100)
center1y = np.random.normal(2 , mu1 , 100)
center2x = np.random.normal(3 , mu1 , 100)
center2y = np.random.normal(5 , mu1 , 100)
center3x = np.random.normal(5 , mu1 , 100)
center3y = np.random.normal(3 , mu1 , 100)

kx1 = np.square(center1x - 2)
ky1 = np.square(center1y - 2)
kx2 = np.square(center2x - 3.5)
ky2 = np.square(center2y - 5)
kx3 = np.square(center3x - 5)
ky3 = np.square(center3y - 3.5)

center1z = (1/(2*math.pi*.25))*np.exp(-(kx1)/(2*.25)) * (1/(2*math.pi*.25))*np.exp(-(ky1)/(2*.25))
center2z = (1/(2*math.pi*.25))*np.exp(-(kx2)/(2*.25)) * (1/(2*math.pi*.25))*np.exp(-(ky2)/(2*.25))
center3z = (1/(2*math.pi*.25))*np.exp(-(kx3)/(2*.25)) * (1/(2*math.pi*.25))*np.exp(-(ky3)/(2*.25))
#center2z = (1/(2*math.pi*1))*np.exp(-(kx2)/(2*1)) * (1/(2*math.pi*1))*np.exp(-(ky2)/(2*1))
#center3z = (1/(2*math.pi*.5625))*np.exp(-(kx3)/(2*.5625)) * (1/(2*math.pi*.5625))*np.exp(-(ky3)/(2*.5625))

x_values = np.append(center1x , [center2x , center3x])
y_values = np.append(center1y , [center2y , center3y])
z_values = np.append(center1z , [center2z , center3z])

```

```python
#Plot the joint distributions

ax = plt.axes(projection='3d')
ax.scatter3D(x_values, y_values, z_values, c=z_values, cmap='Blues')
```

```python
#Create csv file from those values

array = np.stack([x_values, y_values, z_values],axis=-1)

#Shuffle each center within the array

np.random.shuffle(array)

np.savetxt("data.csv", array, delimiter=",")
```
